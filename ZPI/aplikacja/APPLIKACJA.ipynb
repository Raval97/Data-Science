```python
import pandas as pd
import numpy as np
import datetime
import re
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score as acc
import tkinter as tk
from tkinter import font
from tkinter import simpledialog
from tkinter import *
from PIL import ImageTk as itk
from PIL import  Image
import tkinter.font
from sklearn.model_selection import GridSearchCV
from math import sqrt
import sklearn.metrics as sm
from sklearn.neighbors import KNeighborsRegressor
from sklearn.datasets import make_regression
```

```python
data = pd.read_csv("data_result", sep='\t') 
```

```python
data_selected = data.iloc[:, 2:7]
data_selected.describe()
```

```python
data.describe()
```

```python
mins = []
maxs = []
for i in range(0,5):
    mins.append(data_selected.iloc[:,i].min())
    maxs.append(data_selected.iloc[:,i].max())

print(mins, maxs)
```

```python
def bitfield(n):
    return [int(digit) for digit in bin(n)[2:]]

a = bitfield(3)
N = 5
a = [0] * (N - len(a)) + a
a
```

```python
#MODEL DLA DecisionTreeRegressor
select_data_1 = data
Y_2 = select_data_1['dKominSO2']
select_data_1 = select_data_1.drop(['dKominSO2'], axis = 1)
select_data_1 = select_data_1.drop(['Unnamed: 0'], axis = 1)
min_max_scaler = preprocessing.MinMaxScaler()
data_scaled_2 = min_max_scaler.fit_transform(select_data_1)

print(data_scaled_2.mean(axis=0))
print(data_scaled_2.std(axis=0))
x2 = pd.DataFrame(data_scaled_2, columns = [ 'd14HTJ10CL001',  'd14HTA21CT001',  'd14HTF15CF001', 'd14HTA21CP001', 'd14HTA23CT002'])

training_data2, testing_data2, training_target2, testing_target2 = train_test_split(x2, Y_2, test_size=0.3)
clf2 = tree.DecisionTreeRegressor()

clf2 = clf2.fit(training_data2, training_target2)

print("Mean squared error =", round(sm.mean_squared_error(testing_target2, clf2.predict(testing_data2)), 2)) 
print("Median absolute error =", round(sm.median_absolute_error(testing_target2, clf2.predict(testing_data2)), 2)) 
print("Explain variance score =", round(sm.explained_variance_score(testing_target2, clf2.predict(testing_data2)), 2)) 
print("R2 score =", round(sm.r2_score(testing_target2, clf2.predict(testing_data2)), 2))
```

```python
#MODEL KONCOWY 

regr = KNeighborsRegressor()
regr.fit(training_data2, training_target2)


print("Mean squared error =", round(sm.mean_squared_error(testing_target2, regr.predict(testing_data2)), 2)) 
print("Median absolute error =", round(sm.median_absolute_error(testing_target2, regr.predict(testing_data2)), 2)) 
print("Explain variance score =", round(sm.explained_variance_score(testing_target2, regr.predict(testing_data2)), 2)) 
print("R2 score =", round(sm.r2_score(testing_target2, regr.predict(testing_data2)), 2))
```

```python
res = []
for i in range(0, 32):
    pred = []
    a = bitfield(i)
    N = 5
    bitrep = [0] * (N - len(a)) + a
    for j in range(0, N):
        if bitrep[j] == 0:
            pred.append(mins[j])
        else:
            pred.append(maxs[j])
    pred.append(regr.predict(np.array(pred).reshape(1, -1))[0])
    res.append(pred)
    
df = pd.DataFrame(res, index = [x for x in range(1, 33)], columns =["d14HTJ10CL001", "d14HTA21CT001", "d14HTF15CF001", "d14HTA21CP001", "d14HTA23CT002", "dKominSO2"])

df
```
