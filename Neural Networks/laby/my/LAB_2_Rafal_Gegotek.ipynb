```python
import sys
import numpy as np
import math
```

```python
data = np.array([[0.08,0.72,1.0],
        [0.01,1.00,0.0],
        [0.26,0.58,1.0],
        [0.35,0.95,0.0],
        [0.45,0.15,1.0],
        [0.60,0.30,1.0],
        [0.70,0.65,0.0],
        [0.92,0.45,0.0]])
weights = np.array([1.00, -1.00])
#bias = 0.20
```

```python
def sigmoid(x):
    return 1.0 / ( 1.0 + np.exp(-x))
```

```python
def sigmoid_dx(x):
    return sigmoid(x) * (1-sigmoid(x))
```

```python
def loss(prediction, excepted):
    return (prediction-excepted)**2
```

```python
def loss_dy(prediction, excepted):
    return 2*(prediction - excepted)
```

```python
def accuracy(data, weights):
    predictions = sigmoid(np.dot(data[:,:-1], weights))
    predictions = [1.0 if p > 0.5 else 0.0 for p in predictions]
    return np.sum(predictions == data[:,-1])/len(data[:,-1])
```

```python
def train_weights(data, weights, nb_epochs, l_rate):
    treshold = 0.1
    for epoch in range(nb_epochs):
        prediction = sigmoid(np.dot(data[:,:-1], weights))
        loss_score = loss(prediction, data[:, -1]).sum()/data.shape[0]
        gradient = np.dot(data[:, :-1].T, loss_dy(prediction, data[:,-1])*sigmoid_dx(data[:, -1]))
        weights -= l_rate * gradient
        acc_score = accuracy(data, weights)
        print('\nEpoch:',epoch+1,'\nMSE=',loss_score,'\nweights:',weights,'\nAccuracy',acc_score)
#         if treshold > loss_score:
#             break
    return weights

train_weights(data, weights, 1000, 1)
```
